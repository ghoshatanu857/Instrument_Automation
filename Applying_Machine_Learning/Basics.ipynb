```python
# for image manipulation 
import numpy as np
import imageio.v2 as imageio
from imageio.v2 import imread, imsave
from skimage.transform import resize

# Read a JPEG image into a numpy array 
img = imread(r"C:\Users\Basanta_Atanu\OneDrive\Desktop\Atanu\Atanu_Github\Instrument_Automation\Applying_Machine_Learning\Image_basic_Scipy_ImageProcessing.jpg") # path of the image 
# print(img,img.dtype, img.shape) 

# Tinting the image 
img_tint = img * [1, 0.4, 0.3] 
# print(img_tint)

# Saving the tinted image 
imsave(r"C:\Users\Basanta_Atanu\OneDrive\Desktop\Atanu\Atanu_Github\Instrument_Automation\Applying_Machine_Learning\cat_tinted.jpg", img_tint.astype(np.uint8)) 

# Resizing the tinted image to be 300 x 300 pixels 
img_tint_resize = resize(img_tint, (300, 300))

# # Saving the resized tinted image 
imsave(r"C:\Users\Basanta_Atanu\OneDrive\Desktop\Atanu\Atanu_Github\Instrument_Automation\Applying_Machine_Learning\cat_tinted_resized.jpg", img_tint_resize.astype(np.uint8))
```

```python
# Python script using Scikit-learn 
# for Decision Tree Classifier 

# Sample Decision Tree Classifier 
from sklearn import datasets 
from sklearn import metrics 
from sklearn.tree import DecisionTreeClassifier 

# load the iris datasets 
dataset = datasets.load_iris() 
# print(dataset)

# fit a CART model to the data 
model = DecisionTreeClassifier() 
model.fit(dataset.data, dataset.target) 
print(model) 

# make predictions 
expected = dataset.target 
predicted = model.predict(dataset.data) 

# summarize the fit of the model 
print(metrics.classification_report(expected, predicted)) 
print(metrics.confusion_matrix(expected, predicted)) 
```
